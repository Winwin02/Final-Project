```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#data
x5102 = [18,16,20,22,26,12,14,20]
y5102 = [12,10,8,20,24,10,16,18]
```

```python
df5102 = pd.DataFrame(list(zip(x5102,y5102)), columns=['x','y'])
df5102
```

```python
df5102.describe()
```

```python
np.mean(df5102)
```

```python
df5102['x'].mean()
```

```python
x_mean = df5102['x']-df5102['x'].mean()
x_mean
```

```python
y_mean = df5102['y']-df5102['x'].mean()
y_mean
```

```python
sum5102 = np.sum(x_mean*y_mean)
print(sum5102)
```

```python
sqrt5102 = np.sqrt(np.sum(x_mean**2)*np.sum(y_mean**2))
sqrt5102
```

```python
correlation = sum5102/sqrt5102
print(correlation)
```

```python
np.corrcoef(df5102['x'],df5102['y'])
```

```python
plt.title('Korelasi variable x dan y')
plt.scatter(df5102['x'], df5102['y'])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
